#### Remove the table headers of all four pages
#### where a part of the headers is not in the correct place,thus need to be corrected

````python
# collect the index of all table headers and unnecessary area of the tables from all 4 pages
positions = []
for index, line in enumerate(lines):
    if "TABLE" in line:
        header = []
        header.append(index)
    if line.startswith('and areas'):
        header.append(index)
        positions.append(header)
    if line.startswith('SUMMARY'):
        positions.append(index)
    
````

````python
lines[:] = lines[positions[0][1]  : positions[1][0]] +\
           lines[positions[1][1]  : positions[2][0]] +\
           lines[positions[2][1]  : positions[3][0]] +\
           lines[positions[3][1]  : positions[-1]]
````

````python
#collect the index of mis-placed table headers in the following 3 pages and drop them
mis_header = []
for index, line in enumerate(lines):
    if line.startswith('Im'):
        p = []
        p.append(index)
    if line.startswith('2011'):
        p.append(index)
        mis_header.append(p)
````

````python
lines[:] = lines[:mis_header[0][0]] +\
           lines[mis_header[0][1] + 1: mis_header[1][0] ] +\
           lines[mis_header[1][1] + 1: mis_header[2][0] ] +\
           lines[mis_header[2][1] + 1: ]
````

````python
#remaining mis_headers
mis_header2 = []
for index, line in enumerate(lines):
    if line.startswith('Child'):
        p = []
        p.append(index)
    if line.startswith('ITN (%)'):
        p.append(index)
        mis_header2.append(p)
````

````python
#design a regular expression pattern to extract all the country names
p1 = r'^([A-Z])(.+?)\s*\d*\n'
````

````python
finalCountries = []
for c in countriesData:
    m = re.match(p1,c)
    if m:
        finalCountries.append(m.group(1) + m.group(2))
        
````

### 2.3 Extract all columns

#### first fix the the only data of the first column that is falsely merged into the country name

````python
for index, line in enumerate(countriesData):
        if 'Democratic Republic of the Congo' in line:
            p = index
p2 = '^([A-z].+?)\s*(\d*)\n'
match = re.match(p2,countriesData[p])
countriesData = countriesData[:p] + [match.group(2),match.group(1)] + countriesData[p + 1:]
````

````python
recodrsPerpage =[62,63,63,14]
````

#### Now we can start to capture data for all 22 columns

````python
num_of_cols = 22
data = {}
for i in range(num_of_cols):
    data[i] = []
````

````python
pageNum = -1 
numRecs = 0
colIdx = 0
idx = 0
for line in countriesData:
    if line.startswith('and areas'):
        pageNum += 1
        numRecs = recodrsPerpage[pageNum]
        colIdx = 0
        idx = 0
    if re.match(p3,line) and colIdx < num_of_cols:
        data[colIdx].append(re.match(p3,line).group(0))
        idx += 1
        if idx % numRecs == 0:
            colIdx += 1
````

### 2.4 Store data in csv format

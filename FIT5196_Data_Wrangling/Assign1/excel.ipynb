### Load data and store in a DataFrame

```python
df = pd.read_excel('./Input/basic_indicators.xlsx',skiprows=list(range(5)),names=['x','Country Name']+list(range(22)))
```

```python
#insepect the whole dataframe
df
```

```python
#drop the null columns and the unneeded part of the table.
#and make the columns to be continious
df = df.dropna(axis=1,how='all')
df = df.drop(axis=1,columns=[13,15])
df = df[:202]
df = df.rename(columns={14:13})
df
```

```python
#round the float point to integer and convert them to string 
df = df.apply(lambda x : x.apply(lambda elt: str(int(elt)) if isinstance(elt, float) else str(elt)));df
```

```python
#convert all '–' to empty field
df = df.apply(lambda x: x.str.replace('–',''),axis=1);df
```

### Output file

```python
#output the DataFrame to CSV file without adding the numeric indexes as the first column
df.to_csv('./Output/basic_indicators.csv',index=False)
```
